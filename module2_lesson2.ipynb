```python
import numpy as np
from matplotlib import pyplot as pl

%matplotlib inline
```

```python
def linear_convection(nx, L=2.0, c=1.0, dt=0.025, nt = 20):
    dx = L / (nx - 1)
    x = np.linspace(0.0, L, num = nx)
    u0 = np.ones(nx)
    mask = np.where((x>=0.5) & (x<=1.0))
    u0[mask] = 2.0
    u = u0.copy()
    for n in range(nt):
        un = u.copy()
        u[1:] = un[1:] - c * dt / dx * (un[1:] - un[:-1])
    pl.figure()
    pl.grid()
    pl.plot(x, u0, label='initial')
    pl.plot(x, u, label='current')
    pl.legend()
    pl.xlim(0.0, L)
    pl.ylim(0.0, 2.5)
```

```python
linear_convection(41)
```

```python
linear_convection(85)
```

```python
def linear_convection_cfl(nx, L=2.0, c=1.0, sigma=0.5, nt = 20):
    dx = L / (nx - 1)
    x = np.linspace(0.0, L, num = nx)
    dt = sigma * dx / c
    u0 = np.ones(nx)
    mask = np.where((x>=0.5) & (x<=1.0))
    u0[mask] = 2.0
    u = u0.copy()
    for n in range(nt):
        un = u.copy()
        u[1:] = un[1:] - c * dt / dx * (un[1:] - un[:-1])
    pl.figure()
    pl.grid()
    pl.plot(x, u0, label='initial')
    pl.plot(x, u, label='current')
    pl.legend()
    pl.xlim(0.0, L)
    pl.ylim(0.0, 2.5)
```

```python
linear_convection_cfl(85, sigma = 0.9)
```
