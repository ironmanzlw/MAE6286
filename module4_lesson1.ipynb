```python
import numpy as np
from matplotlib import pyplot as pl

%matplotlib inline
```

```python
def ftcs (u0, alpha, dt, dx, nt):
    u = u0.copy()
    for n in range (nt):
        u[1:-1] = u[1:-1] +  sigma * (u[:-2] - 2 * u[1:-1] + u[2:])
        #update value
        u[0] = u0[0]
        u[-1] = u0[-1]
    
    return u
```

```python
#paramater
L = 1.0
nx = 51
dx = L/(nx-1)
sigma = 1/2
alpha =  1.22*10**(-3)
dt = sigma * dx**2 / alpha

# intitial condition
x = np.linspace(0.0, L, num = nx)
T0 = np.zeros(nx)
T0[0] = 100.0
T0[-1] = 0
```

```python
nt = 100
T100 = ftcs(T0, alpha, dt, dx, nt)
pl.plot(x, T100)
#pl.plot(x, T100_2)
```

```python
nt = 1000
T1000 = ftcs(T0, alpha, dt, dx, nt)
pl.plot(x,T1000)
```

```python
def ftcs2 (u0, alpha, dt, dx, nt, q = 0.0):
    u = u0.copy()
    for n in range (nt):
        u[1:-1] = u[1:-1] +  sigma * (u[:-2] - 2 * u[1:-1] + u[2:])
        #update left (dirichlet)
        u[0] = u0[0]
        #update right (neumann)
        u[-1] = u[-2] + q*dt
    
    return u
```

```python
nt = 100
T100 = ftcs2(T0, alpha, dt, dx, nt)
pl.plot(x, T100)
#pl.plot(x, T100_2)
```

```python
nt = 1000
T1000 = ftcs2(T0, alpha, dt, dx, nt)
pl.plot(x,T1000)
```
